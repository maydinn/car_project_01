```python
#func >>> from string to float
def string_to_float(x):
    a=""
    if isinstance(x, (str)) :
        for i in x:
            if i.isnumeric():
                a += i
        if a.isnumeric():
            return a
        elif a == "":
            return np.NaN
        else:
            raise Exception("Sorry, there shoul be someting wrong with for loop")
    elif isinstance(x, (int, float)):
        return x
    elif x == None:
        return np.NaN
    elif np.isnan(x):
        return np.NaN
    else:
        raise Exception("Sorry, all values should be string, float or int")
    
```

```python
def list_to_String(x):
    a = []
    if (isinstance(x, (list))):
        for i in x:
            if (isinstance(i, (float, int, str))) and (i != "") and (i != "\n"):
                          a.append(i) 
    elif np.isnan(x):
        return np.NaN
    else:
        raise Exception ("Sorry, some values are not list")
    a= " ".join(a)
    return re.sub('\n', "", a)
```
