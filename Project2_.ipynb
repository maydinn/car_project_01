```python
def unique_values(data, col_name):
    uniqule_values = set()
    for rows in data[col_name]:
        uniqule_values.update(rows)
    return list(uniqule_values)
```

```python
def nan_to_list(data, column_name):
    nan_rows=data[column_name].isnull()
    data.loc[nan_rows, column_name] = data.loc[nan_rows, column_name].apply(lambda x: [])
```

```python
def fill_na(x, group_name, col_name):
    first= x[group_name].unique()
    c= col_name
    for i in first:
        liste = list(x[(x['make_model']==i)].index)
        a=x[(x['make_model']==i)].mode()
        if a.shape[0] > 0:
            a=a.iloc[0]
            for t in liste:
                if pd.isna(x.loc[t, c]):
                    x.loc[t,c] = a[c]
            else:
                continue
                
```
